```python
cursor = mydb.cursor()
```

```python
import xlrd
```

```python
loc = xlrd.open_workbook("practice_data.xls")
sheet = loc.sheet_by_name("Dept. FOIA Contacts")
```

```python
cursor.execute("create table car_dataset.practice(agency VARCHAR(200), department VARCHAR(200), name VARCHAR(200), room_number VARCHAR(200),street_address VARCHAR(200), city VARCHAR(200), state VARCHAR(200), zip_code VARCHAR(200),telephone VARCHAR(200), fax VARCHAR(200), email_address VARCHAR(200),website VARCHAR(200), online_request VARCHAR(250), notes longtext)")
```

```python
mydb.commit()
```

```python
q= """insert into practice(agency, department, name,room_number, street_address, city, state, zip_code, telephone,fax, email_address, website, online_request,notes) values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
#here, only a query is written with placeholders.
```

```python
#Here, values are assigned to each of the columns from the excel sheet.
for i in range(1, sheet.nrows):
    agency = sheet.cell(i,0).value #adding .value converts it to a string.
    department = sheet.cell(i,1).value
    name = sheet.cell(i,2).value
    room_number = sheet.cell(i,4).value
    street_address = sheet.cell(i,5).value
    city = sheet.cell(i,6).value
    state = sheet.cell(i,7).value
    zip_code = sheet.cell(i,8).value
    telephone = sheet.cell(i,9).value
    fax = sheet.cell(i,10).value
    email_address = sheet.cell(i,11).value
    website = sheet.cell(i,12).value
    online_request = sheet.cell(i,13).value
    notes = sheet.cell(i,14).value
    #here, the placeholders are given the values by which they have to be filled.
    values = (agency, department, name,room_number, street_address, city, state, zip_code, telephone, fax, email_address, website, online_request, notes)
    #Query is executed here.
    cursor.execute(q,values)
```
